```python
## Importar librerias 

import re
from bs4 import BeautifulSoup
import requests
```

```python
# se define variable con url sólo por orden

url = 'https://www.dafiti.cl/descuentos-entre-50-y-70/'

# obtención de la página que se va a analizar

page = requests.get(url)

print(f"Codigo: {page.status_code}")

# el contenido de la pagina se encuentra en page.content, y con esto se lee con BeautifulSoap

page = BeautifulSoup(page.content, 'html.parser')

print(page.prettify())

```

```python
all_products = page.find_all(class_='itm-link')

print(len(all_products))
```

```python
for product in all_products:

    print('=============================================================')

    marca_nombre = product.find(class_='card-description').text

    print(marca_nombre)

    precio = product.find(class_='discount-price').text.replace(' ','').replace('$','').replace('.','')

    print('Precio de rebaja = $', precio)

    precio_normal = product.find(class_='price-prefix-listing special itm-price special original-price').text.replace(' ','').replace('$','').replace('.','')

    print('Precio anterior = $', int(precio_normal))

    descuento = product.find(class_='card-footer card-footer-ab').text.replace('-','').replace(' ','').replace('%','')

    print('Descuento = ', int(descuento), '%')


    envio = product.find(class_='shipping-text')
    

    if envio is None:
        print("no ahi envio gratis")
    else:
        print("si ahi envio gratis")

    ahorro = (int(precio_normal)-(int(precio)))

    print('Ahorro = $',ahorro)


    Cuotas = product.find(class_='itm-price-installments footer-installments').text.replace(' ','_')

    print(Cuotas)

    preciolist = list[precio]
```
